```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def create_square(shape: np.ndarray = 64, min_size: int = 10, max_size: int = 20, fixed_size: int = None):

    assert min_size <= max_size, "min_size must be smaller than max_size"
    assert shape > max_size, "shape must be larger than max_size"
    assert shape > min_size, "shape must be larger than min_size"

    if fixed_size:
        # Set square size
        square_size = fixed_size
    else:
        # Create a randomly sized square at a random position
        square_size = np.random.randint(min_size, max_size)

    image = np.zeros((shape, shape))

    x = np.random.randint(0, shape - square_size)
    y = np.random.randint(0, shape - square_size)

    image[x : x + square_size, y : y + square_size] = 1

    return image


def create_circle(shape: np.ndarray = 64, min_size: int = 10, max_size: int = 20, fixed_size: int = None):

    assert min_size <= max_size, "min_size must be smaller than max_size"
    assert shape > max_size, "shape must be larger than max_size"
    assert shape > min_size, "shape must be larger than min_size"

    if fixed_size:
        # Set circle size
        radius = fixed_size // 2
    else:
        # Create a randomly sized circle at a random position
        radius = np.random.randint(min_size, max_size) // 2

    image = np.zeros((shape, shape))

    x = np.random.randint(0, shape - radius * 2)
    y = np.random.randint(0, shape - radius * 2)

    for i in range(shape):
        for j in range(shape):
            if (i - radius - x) ** 2 + (j - radius - y) ** 2 < (radius) ** 2:
                image[i, j] = 1

    return image
```

```python
# Try the functions
image = create_square(shape=64, min_size=30, max_size=60)
plt.imshow(image, cmap="gray")
plt.show()

image = create_circle(shape=64, min_size=30, max_size=60)
plt.imshow(image, cmap="gray")
plt.show()
```
